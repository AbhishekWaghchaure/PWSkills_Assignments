## Question 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot


## Answer :

### Five plots in Seaborn along with their usage is as following :

1. **Box Plot** : A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. It is used to identify the spread and skewness of the data.
2. **Heatmap** : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.
3. **Pair Plot** : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between them.
4. **Violin Plot** : A violin plot is a graphical representation of data that combines elements of a box plot and a kernel density plot. It is used to visualize the distribution of a dataset and show its density.
5. **Joint Plot** : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

## Question 2 : Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
### Note: timepoint, signal, event, and region are columns in the fmri dataset.


## Answer :

```python
import seaborn as sns
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
fmri.info()
```

```python
fmri['event'].value_counts()
```

```python
fmri['region'].value_counts()
```

### Using sns.lineplot() to plot the data

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'event', style='region')
plt.title('Signal over Time for Differnet Events and Regions')
```

## Question 6 : Use the "flights" dataset from seaborn to plot a heatmap
---

```python
flights = sns.load_dataset('flights')
```

```python
print(flights.info())
print(flights.head())
```

### Pivot the dataset to create a matrix with month on the x-axis, year on the y-axis, and passengers as values

```python
flight_pivot = flights.pivot(index = 'year',columns = 'month', values = 'passengers')
```

```python
flight_pivot.corr
```

```python
sns.heatmap(flight_pivot,annot =True,linewidth = .5,fmt = 'd',linecolor='black',cmap='YlGnBu')
plt.title('Passengers per month')
plt.show()
```
